```python
data = pd.read_csv('Orange_Telecom_Churn_Data.csv')
```

```python
data['voice_mail_plan']=data['voice_mail_plan'].map({'yes':1,
                              'no':0})
required_features = ['account_length','voice_mail_plan','number_vmail_messages','total_day_minutes','total_night_minutes','number_customer_service_calls']
```

```python
train_data = data[required_features]
```

```python
train_data.head()
```

```python
test_data  = data.iloc[:,20]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
x , x_test ,y , y_test = train_test_split(train_data,test_data,random_state = 0,test_size=0.2)
```

```python
classifier = KNeighborsClassifier(n_neighbors = 10 , p=2, metric ='euclidean'  )
```

```python
classifier.fit(x,y)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cn   = confusion_matrix(y_test,y_pred)
```

```python
cn
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```
